```python
import re
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
corpus=[]
```

```python
lm=WordNetLemmatizer()
```

```python
for i in range(len(messages)):
    sms=re.sub('[^a-zA-Z]',' ',messages['message'][i])   #Replace all the thing with " " except it is alphabet
    sms=sms.lower()  #Lowering Case
    sms=sms.split()  #Splitting Sentence
    sms=[lm.lemmatize(word) for word in sms if word not in set(stopwords.words('english'))] # Lemmitization of Sentances
    sms=' '.join(sms)  # Join the Sentance
    corpus.append(sms)  # Appending in List Corpus
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=3000,ngram_range=(1,3))
X=cv.fit_transform(corpus).toarray()
```
